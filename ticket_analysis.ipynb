```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset into a Pandas DataFrame
df=pd.read_csv('economy.csv')
```

```python
# Expl0re data
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# convet date column to datetime
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
```

```python
# extract day of week in a seperate column
df['day'] = df['date'].dt.day_name()
```

```python
# remove commas from price
df['price'] = df['price'].str.replace(',', '').astype(int)
```

```python
# summary stats
df.describe()
```

```python
# explore data with changes
df.head()
```

```python
df.tail()
```

```python
# Average price per day
price_per_day = df.groupby('day')['price'].mean().sort_values()
print(price_per_day)
```

```python
# Average price per airline
avg_airline_price = df.groupby('airline')['price'].mean().sort_values()
print(avg_airline_price)
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='day', y='price', data=df, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Ticket Prices by day')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='day', y='price', data=df, estimator='mean', errorbar=None)
plt.title('Average Ticket Price Per Day')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='day', y='price', hue='airline', data=df)
plt.title('Comparison of Average Prices Across Airlines by Day')
plt.legend(title='Airline', bbox_to_anchor=(1, 1))
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Ticket Prices')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='day', y='price', hue='airline', estimator='mean', errorbar=None)
plt.title('Average Ticket Prices of Each Airline Per Day')
plt.legend(title='Airline', bbox_to_anchor=(1, 1))
plt.show()
```

```python
cheapest_day = price_per_day.idxmin()
most_expensive = price_per_day.idxmax()
cheapest_airline = avg_airline_price.idxmin()
expensive_airline = avg_airline_price.idxmax()
```

```python
print(f"Cheapest day to book flights: {cheapest_day}")
print(f"Most expensive day to book flights: {most_expensive}")
print(f"Cheapest airline: {cheapest_airline}")
print(f"Most expensive airline: {expensive_airline}")
```
